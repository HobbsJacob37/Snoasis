```python
import pandas as pd
import matplotlib.pyplot as plt
import calendar
day_names = [calendar.day_name[i] for i in range(0, 7)]

# Load the data into a dataframe
df = pd.read_csv('https://raw.githubusercontent.com/camdenlol/SnOasis/main/snoasisData.csv')

# Convert the 'Date' column to a datetime object
df['Date'] = pd.to_datetime(df['Date'])

# Sort the dataframe by the 'Date' column
df = df.sort_values(by='Date')

# Extract the day from the 'Date' column
df['day'] = df['Date'].dt.day

# Create a histogram of the 'day' column
plt.hist(df['day'], bins=range(1, 9), align='left', rwidth=0.8)

# Set the x-axis tick labels to the day names
plt.xticks(range(1, 8), day_names, rotation=45, ha='right')

# Customize the plot
plt.title('Bar Chart of Data by Day')
plt.xlabel('Day')
plt.ylabel('Sales')

# Display the plot
plt.show()
```
